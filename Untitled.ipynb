```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import textdistance as txt
```

```python
df = pd.read_csv('search.csv', skiprows=6)
df.head()
```

```python
df.info()
```

```python
df['Termo de pesquisa'] = df['Termo de pesquisa'].str.lower()
df = df[['Termo de pesquisa','Total de pesquisas exclusivas']]
df = df.set_index('Termo de pesquisa')
```

```python
df = df[df['Total de pesquisas exclusivas'].apply(lambda x: x.isnumeric())]
```

```python
df.columns = ['pesquisas']
df['pesquisas'] = pd.to_numeric(df.pesquisas)
df.info()
```

```python
df.index = df.index.str.replace('í','i')
df
```

```python
df = df.groupby(['Termo de pesquisa']).sum()
```

```python
df = df.sort_values(by=['pesquisas'],ascending=False)
```

```python
df = df.drop(['rj','rio de janeiro','concurso'], axis=0)
df.head(50)
```
